### Importación de datos



```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.graph_objects as go

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

# 1. Análisis de facturación



```python


ts = list(enumerate([tienda, tienda2, tienda3, tienda4], 1))
facturacion = [sum(t[1]['Precio']) for t in ts]
tiendas = [f"Tienda {t[0]}" for t in ts]

for t in (ts := list(enumerate([tienda, tienda2, tienda3, tienda4], 1))):
    print(f"Tienda {t[0]}: $ {sum(t[1]['Precio']):,}")

plt.figure(figsize=(8, 6))
plt.bar(tiendas, facturacion, color='skyblue')
plt.title('Facturación Total por Tienda')
plt.xlabel('Tienda')
plt.ylabel('Facturación ($)')
plt.show()
```

# 2. Ventas por categoría

```python
for t in ts:
    print(f"Tienda {t[0]}")
    ventas_por_categoria = (
        t[1].groupby(['Categoría del Producto'])  # Agrupa por tienda y categoría
            .size()                                        # Cuenta registros (ventas) por grupo
            .sort_values(ascending=False)                  # Ordena de mayor a menor
            .reset_index(name='Número de Ventas')          # Convierte a DataFrame y renombra columna
    )

    print(ventas_por_categoria)

    df = t[1].groupby('Categoría del Producto')['Producto'].value_counts()
    print(df)

    plt.figure(figsize=(10, 6))
    ventas_por_categoria.plot(kind="bar", title=f"Ventas por Categoría en Tienda {t[0]}")
    plt.xlabel('Categoría del Producto')
    plt.ylabel('Número de Ventas')
    plt.xticks(rotation=45, ha='right')
    plt.show()
```

# 3. Calificación promedio de la tienda


```python
for t in ts:
    print(f"Tienda {t[0]}: {t[1]['Calificación'].mean()}")

fig, ax1 = plt.subplots(figsize=(8, 6))

ax1.bar(tiendas, facturacion, color='skyblue', label='Facturación')
ax1.set_xlabel('Tienda')
ax1.set_ylabel('Facturación ($)', color='skyblue')
ax1.tick_params(axis='y', labelcolor='skyblue')

ax2 = ax1.twinx()
calificaciones = [t[1]['Calificación'].mean() for t in ts]
ax2.plot(tiendas, calificaciones, color='red', marker='o', label='Calificación Promedio')
ax2.set_ylabel('Calificación Promedio', color='red')
ax2.tick_params(axis='y', labelcolor='red')

plt.title('Facturación y Calificación Promedio por Tienda')
fig.tight_layout()
plt.show()
```

# 4. Productos más y menos vendidos

```python
import matplotlib.pyplot as plt
for t in ts:
    print(f"Tienda {t[0]}:\n{(cp := t[1]['Producto'].value_counts())}")
    plt.figure(figsize=(10, 8))
    cp.sort_values().plot(kind="barh", title=f"Tienda {t[0]}")
    plt.show()
```

# 5. Envío promedio por tienda

```python
envios_promedio = []
for t in ts:
    print(f"Tienda {t[0]}: {(envio_promedio := t[1]['Costo de envío'].mean()):,}")
    envios_promedio.append(envio_promedio)
plt.figure(figsize=(8, 6))
plt.pie(envios_promedio, labels=tiendas, autopct='%1.1f%%', startangle=90)
plt.title('Costo de Envío Promedio por Tienda')
plt.xlabel('Tienda')
plt.show()
```

```python
# Lista de colores para cada tienda
colores = ['red', 'blue', 'green', 'orange']  # Agrega más colores si es necesario

# Crear el gráfico 3D interactivo con leyenda individual
fig = go.Figure()

for i in range(len(tiendas)):
    fig.add_trace(go.Scatter3d(
        x=[facturacion[i]],
        y=[calificaciones[i]],
        z=[envios_promedio[i]],
        mode='markers',
        marker=dict(
            size=12,
            color=colores[i],  # Asignar color de la lista
            opacity=0.8
        ),
        name=tiendas[i],  # Nombre de la tienda para la leyenda
        text=[tiendas[i]]
    ))

# Configurar el diseño del gráfico
fig.update_layout(
    scene=dict(
        xaxis_title='Facturación',
        yaxis_title='Calificación Promedio',
        zaxis_title='Costo de Envío Promedio'
    ),
    title='Análisis de Tiendas en 3D',
    showlegend=True  # Mostrar la leyenda
)

fig.show()
```
